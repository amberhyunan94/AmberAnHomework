```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

school_copy = school_data.copy()

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
school_df = pd.DataFrame(school_data_complete)
school_df.head()
```

```python
schools = school_df['school_name'].unique()
total_schools = len(schools)
total_students = school_df['student_name'].count()
budgets = school_df['budget'].unique()
total_budget = budgets.sum()
avg_math = school_df['math_score'].mean()
avg_reading = school_df['reading_score'].mean()
passing_math_df = school_df.loc[school_df['math_score']> 69]
passing_reading_df = school_df.loc[school_df['reading_score']> 69]
percent_passing_math = len(passing_math_df)/total_students
percent_passing_reading = len(passing_reading_df)/total_students
overall_passing_rate = (percent_passing_math+percent_passing_reading)/2

District_Summary = pd.DataFrame({
    "Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Math Score": [avg_math],
    "Average Reading Score": [avg_reading],
    "% Passing Math": [percent_passing_math],
    "% Passing Reading":[percent_passing_reading],
    "Overall Passing Rate":[overall_passing_rate]
})
District_Summary = District_Summary.style.format({"Total Budget": "${:,.2f}",
                      "% Passing Math": "{:.1%}","% Passing Reading": "{:.1%}","Overall Passing Rate": "{:.1%}"})
District_Summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
del school_copy['School ID']
school_copy
```

```python
school_copy['Per Student Budget'] = school_copy['budget']/school_copy['size']
avg_math_reading_scores = school_df.groupby(['school_name'])['reading_score','math_score'].mean().reset_index()
school_copy = school_copy.merge(avg_math_reading_scores,on='school_name',how='outer')
percent_passing_read = student_data.loc[student_data['reading_score']>=70]
passing_read_count = percent_passing_read.groupby(['school_name'])['reading_score'].count().reset_index()
passing_read_count.rename({'reading_score' : 'Reading Count'}, axis=1, inplace=True)

percent_passing_math = student_data.loc[student_data['math_score']>=70]
passing_math_count = percent_passing_math.groupby(['school_name'])['math_score'].count().reset_index()
passing_math_count.rename({'math_score' : 'Math Count'}, axis=1, inplace=True)
pass_count = passing_math_count.merge(passing_read_count, on="school_name", how='inner')
school_copy = school_copy.merge(pass_count, on="school_name", how='outer')
school_copy
```

```python
school_copy['% Passing Math'] = school_copy['Math Count']/school_copy['size']
school_copy['% Passing Reading'] = school_copy['Reading Count']/school_copy['size']
school_copy['% Passing Overall'] = (school_copy['% Passing Math'] + school_copy['% Passing Reading'])/2
del school_copy['Math Count']
del school_copy['Reading Count']
```

```python
school_copy.rename({'reading_score':'Avg. Reading Score','math_score':'Avg. Math Score','budget':'Total Budget','type':'Type','size':'Size','school_name':'Schools'}, axis=1, inplace=True)
school_copy = school_copy.style.format({"Total Budget": "${:,.2f}",
                      "% Passing Math": "{:.1%}","% Passing Reading": "{:.1%}","% Passing Overall": "{:.1%}"})
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
spend_by_grade = school_copy.data
type(spend_by_grade)
```

```python
binned_spending = pd.cut(spend_by_grade['Per Student Budget'], spending_bins, labels=group_names)
binned_spending_df = pd.DataFrame(binned_spending)
spend_by_grade['Spending Level'] = binned_spending_df
```

```python
scores_by_spending = spend_by_grade.groupby(['Spending Level'])['Avg. Reading Score',
                                                         'Avg. Math Score',
                                                         '% Passing Reading',
                                                         '% Passing Math',
                                                         '% Passing Overall'
                                                         ].mean()
```

```python
scores_by_spending
```

```python
Reformatted_bySpend = scores_by_spending.style.format({"% Passing Math": "{:.1%}","% Passing Reading": "{:.1%}","% Passing Overall": "{:.1%}"})
Reformatted_bySpend
```
