## Headlines & Article Blurbs

```python
#import dependencies
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
```

```python
# read NASA news site with requests
url = 'https://mars.nasa.gov/news/'
response = requests.get(url)
```

```python
#reading headline news
soup = bs(response.text, 'lxml')
```

```python
headlines = soup.find('div', class_ = 'content_title').text.strip()
print(headlines)
```

```python
news = soup.find('div', class_ = 'rollover_description_inner').text.strip()
print(news)
```

```python
print('News Title:',headlines)
print('News Paragraph:',news)
```

## Hemispheres


```python
browser = Browser('chrome', headless=False)
usgs_url = 'https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars'
browser.visit(usgs_url)
usgs_html = browser.html
usgs_soup = bs(usgs_html, 'html.parser')
hemisphere_image_list = []
```

```python
results = usgs_soup.find('div', class_='result-list')
hemisphere_items = results.find_all('div', class_='item')
1
# iterates through hemispheres
for hemisphere in hemisphere_items:
    title = hemisphere.find('div', class_='description')
    title_text = title.a.text
    title_text = title_text.replace(' Enhanced', '')
    # clicking into each of the hemispheres 
    browser.links.find_by_partial_text(title_text)
    usgs_html = browser.html
    usgs_soup = bs(usgs_html, 'html.parser')
    image = usgs_soup.find('div', class_='downloads').find('ul').find('li')
    img_url = image.a['href']
    hemisphere_image_list.append({'title': title_text, 'img_url': img_url})
```
