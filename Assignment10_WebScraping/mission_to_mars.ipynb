## Headline & Article Blurb

```python
#import dependencies
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
```

```python
# read NASA news site with requests
url = 'https://mars.nasa.gov/news/'
response = requests.get(url)
```

```python
#reading headline news
soup = bs(response.text, 'lxml')
```

```python
headlines = soup.find('div', class_ = 'content_title').text.strip()
print(headlines)
```

```python
news = soup.find('div', class_ = 'rollover_description_inner').text.strip()
print(news)
```

```python
print('News Title:',headlines)
print('News Paragraph:',news)
```

```python
#explore jpl site
executable_path = {"executable_path": "chromedriver"}
browser = Browser("chrome", **executable_path, headless=False)
imageurl = 'https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars'
browser.visit(imageurl)
```
